```python
# Compute connectivity matrices for each subject and condition
conns = []
for subj in results["attractive"].keys():
    conn_cond1 = np.array(results['attractive'][subj])
    conn_cond2 = np.array(results['unattractive'][subj])
    conns.append([conn_cond1, conn_cond2])

conns_array = np.array(conns)
# Perform permutation cluster test on connectivity matrices
t_obs, clusters, p_values, _ = permutation_cluster_test(
    conns_array, n_permutations=1000, threshold=None, tail=0)

# Print significant clusters
for i_c, c in enumerate(clusters):
    if p_values[i_c] < 0.05:
        print(f"Cluster {i_c} is significant with p-value {p_values[i_c]}")
```

```python
# Find intersection of column and index labels across all participants
common_cols = set(results['attractive']["p1"].columns)
common_idxs = set(results['attractive']["p1"].index)
for cond in ['attractive', 'unattractive']:
    for subj in results["attractive"].keys():
        common_cols = common_cols.intersection(set(results[cond][subj].columns))
        common_idxs = common_idxs.intersection(set(results[cond][subj].index))

# Create new dictionary with only common columns and indexes
common_data = {}
for cond in ['attractive', 'unattractive']:
    common_data[cond] = {}
    for subj in results["attractive"].keys():
        conns = results[cond][subj]
        conns_common = conns.loc[common_idxs, common_cols]
        common_data[cond][subj] = conns_common
```

```python
# Convert list of connectivity matrices to NumPy array
conns = []
for subj in results["attractive"].keys():
    conn_cond1 = common_data['attractive'][subj]
    conn_cond2 = common_data['unattractive'][subj]

    conns.append([conn_cond1, conn_cond2])

# Convert list of connectivity matrices to NumPy array
conns_array = np.array(conns)

# Perform permutation cluster test on connectivity matrices
t_obs, clusters, p_values, _ = permutation_cluster_test(
    conns_array, n_permutations=1000, threshold=None, tail=0)

# Print significant clusters
for i_c, c in enumerate(clusters):
    if p_values[i_c] < 0.05:
        print(f"Cluster {i_c} is significant with p-value {p_values[i_c]}")
```
